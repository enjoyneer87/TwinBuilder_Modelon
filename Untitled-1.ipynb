```python
import math
import matplotlib.pyplot as plt

#  ()
T_HIGH_C = 60.0
T_LOW_C  = -40.0
T_AMB_C  = 22.0
DWELL_HIGH = 21600    # 6h
DWELL_LOW  = 21600    # 6h
RAMP = 300*6            # 5m
N_CYCLES = 5
AMBIENT_HOLD = 86400  # 24h
RAMP_STEP = 30        #   (30s: 10 points).  : 10,15 

def half_cosine_ramp(t0, T1, T2, Tramp, step):
    """  0 half-cosine  (t0~t0+Tramp).  ."""
    pts = []
    n = int(Tramp/step)
    for i in range(n+1):
        tau = i*step/Tramp
        T = T1 + (T2 - T1)/2 * (1 - math.cos(math.pi * tau))
        pts.append((t0 + i*step, T))
    #    ( step  )
    if pts[-1][0] < t0 + Tramp:
        T = T2
        pts.append((t0 + Tramp, T))
    else:
        #    
        pts[-1] = (t0 + Tramp, T2)
    return pts

times = []
temps_c = []

t = 0.0
for cyc in range(1, N_CYCLES+1):
    # High dwell
    times.append(t); temps_c.append(T_HIGH_C)
    times.append(t + DWELL_HIGH); temps_c.append(T_HIGH_C)
    t += DWELL_HIGH
    # High->Low ramp
    ramp_pts = half_cosine_ramp(t, T_HIGH_C, T_LOW_C, RAMP, RAMP_STEP)
    for tt, Tc in ramp_pts[1:]:  #     t
        times.append(tt); temps_c.append(Tc)
    t += RAMP
    # Low dwell
    times.append(t); temps_c.append(T_LOW_C)
    times.append(t + DWELL_LOW); temps_c.append(T_LOW_C)
    t += DWELL_LOW
    if cyc < N_CYCLES:
        # Low->High ramp
        ramp_pts = half_cosine_ramp(t, T_LOW_C, T_HIGH_C, RAMP, RAMP_STEP)
        for tt, Tc in ramp_pts[1:]:
            times.append(tt); temps_c.append(Tc)
        t += RAMP
    else:
        # : Low->Ambient ramp
        ramp_pts = half_cosine_ramp(t, T_LOW_C, T_AMB_C, RAMP, RAMP_STEP)
        for tt, Tc in ramp_pts[1:]:
            times.append(tt); temps_c.append(Tc)
        t += RAMP
        # Ambient dwell
        times.append(t); temps_c.append(T_AMB_C)
        times.append(t + AMBIENT_HOLD); temps_c.append(T_AMB_C)
        t += AMBIENT_HOLD

# Kelvin 
temps_k = [c + 273.15 for c in temps_c]

# 
plt.figure(figsize=(11,4))
plt.plot(times, temps_c, '-', lw=1.2, label='Half-cosine   (C)')
plt.xlabel(' [s]')
plt.ylabel(' [C]')
plt.grid(alpha=0.3)
plt.title('Thermal Cycling (Half-cosine Ramps)')
plt.legend(loc='upper right')

#  Kelvin
ax = plt.gca()
ax2 = ax.twinx()
ax2.set_ylabel(' [K]')
ymin, ymax = ax.get_ylim()
ax2.set_ylim(ymin + 273.15, ymax + 273.15)

#   
t_cycle_starts = []
t_tmp = 0
for cyc in range(1, N_CYCLES+1):
    t_cycle_starts.append(t_tmp)
    #   ( )
    if cyc < N_CYCLES:
        cycle_len = DWELL_HIGH + RAMP + DWELL_LOW + RAMP
    else:
        cycle_len = DWELL_HIGH + RAMP + DWELL_LOW + RAMP + AMBIENT_HOLD
    t_tmp += cycle_len

for i, cs in enumerate(t_cycle_starts, start=1):
    ax.axvline(cs, color='gray', ls='--', lw=0.5)
    ax.text(cs+600, ymin + (ymax-ymin)*0.05, f'Cycle {i}', fontsize=8, color='gray')

plt.tight_layout()
plt.show()

print(f"   : {t} s = {t/3600:.2f} h")

# () Modelica TimeTable    
def to_modelica_table(times, temps_k):
    rows = []
    for tt, Tk in zip(times, temps_k):
        rows.append(f"  {int(round(tt))}, {Tk:.2f};")
    return "table = [\n" + "\n".join(rows) + "\n];"

#     
# print(to_modelica_table(times, temps_k))
```
